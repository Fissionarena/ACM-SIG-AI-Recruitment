```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Neeraj Nair R\Downloads\StudentsPerformance.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.columns
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['math score'], kde=True)
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Count")
plt.show()
```

```python
#Insight:
#The distribution of math scores shows how student performance is spread.
#Any skewness indicates whether most students score toward lower or higher ranges.
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='reading score', y='writing score', data=df)
plt.title("Reading Score vs Writing Score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
```

```python
#Insight:
#There is a strong positive correlation between reading and writing scores.
#Students who perform well in reading also tend to perform well in writing.
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='gender', y='math score', data=df)
plt.title("Math Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()
```

```python
#Insight:
#This plot helps examine gender-based performance differences.
#Any consistent difference may indicate potential bias or trends.
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title("Math Score by Test Preparation Course")
plt.xlabel("Test Preparation Course")
plt.ylabel("Math Score")
plt.show()
```

```python
#Insight:
#Students who completed the test preparation course generally score higher,
#suggesting preparation positively impacts performance.
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(
    df[['math score', 'reading score', 'writing score']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Between Subject Scores")
plt.show()
```

```python
#Insight:
#Strong correlations exist between all subject scores,
#indicating consistent academic performance across subjects.
```
